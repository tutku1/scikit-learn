```python

#Load dataset

from sklearn.datasets import load_iris
dataSet = load_iris()

features = dataSet.data
labels = dataSet.target
labelsNames = list(dataSet.target_names)
featureNames = dataSet.feature_names

print([labelsNames[i] for i in labels[47:52]])
print(featureNames)


#analyze data
import pandas as pd

#print(type(features))

featuresDF = pd.DataFrame(features)
featuresDF.columns = featureNames

#print(type(featuresDF))
#print(featuresDF.describe()) #veri hakkında bilgi verdi
print(featuresDF.info()) #her bir kolonda boş olmayan kaç satır var ve missing data var mı onu görüntüleyebiliriz

#Visualize data | veriyi görselleştirme

featuresDF.plot(x="sepal length (cm)", y="sepal width (cm)", kind="scatter")


#Select Model | Model seçimi

from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=8)

#veriyi parçalamak lazım | split dataset
import numpy as np
from sklearn.model_selection import train_test_split

X = features
y = labels

X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42) #test olarak verinin %33'ünü ayır

print(y_test[:3])


#Train model | modeli eğitme

clf.fit(X_train, y_train)
accuracy = clf.score(X_train, y_train)
print("accuracy on train data {:.2}%".format(accuracy))


# test model | modeli test edicez

accuracy = clf.score(X_test, y_test)
print("accuracy on test data {:.2}%".format(accuracy))
```
